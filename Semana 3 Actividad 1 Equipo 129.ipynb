Se importan las librerías que vamos a utilizar

Se crea un Data frame a partir de la información obtenida

Se verifica si existen valores nulos en las columnas de la base de datos

```python
df.isnull().any()
```

Como podemos observar, Todas las columnas a excepcion de la "ID" y "X1" contienen por lomenos un valor nulo

Observemos cuantos son los espación vacíos en nuestra base de datos

```python
df_null = df[df.isnull()]
total_registros = len(df)
registros_null = len(df_null)
print("Número total de registro: " + str(total_registros))
print("Total de registros con un campo null: "+ str(registros_null))
```

Esto quiere decir que Todos los registros tienen por lo menos un campo en null. Por tanto, no podemos utilizar la estratégia de eliminar los registros o las Columnas con valores nulos.

Vamos a intentar la estratégia donde se eliminan registros donde todos los valores son "null"

```python
tempDF = df.copy() #Se hace una copia del DataFarme
print("Total de Registros: "+ str(len(tempDF)))
tempDF.dropna(inplace = True) #Se eliminan los registros que son nulos
print("Total de Registros después de eliminar los nulos: "+ str(len(tempDF)))
tempDF.isnull().any() #Se checa si el dataframe aún tiene registros nulos
```

Vamos a probar eliminando registros que tengan más de dos campos nulos

```python
tempDF = df.copy() #Se hace una copia del DataFarme
print("Total de Registros: "+ str(len(tempDF)))
tempDF.dropna(thresh = 23 , inplace = True) #Se eliminan los registros que son nulos
print("Total de Registros después de eliminar los nulos: "+ str(len(tempDF)))
tempDF.isnull().any() #Se checa si el dataframe aún tiene registros nulos
```

Con esta estratégia conservamos 30 registros más que con la anterior. Sin embargo, seguimos teniendo valores nulos en nuestro dataframe

Ahora, vamos a intentar rellenar los valores nulos en nuestro DataFrame

A partir de nuestra descripción de la información podemos tratar rellenar la información de la siguiente manera:

X1,X5-X23: Promedio

X2,X3,X4: Moda


```python
tempDF = df.copy() #Se hace una copia del DataFarme
print("Total de Registros: "+ str(len(tempDF)))

tempDF['X5'].fillna(value = tempDF["X5"].mean(), inplace = True)
tempDF['X6'].fillna(value = tempDF["X6"].mean(), inplace = True)
tempDF['X7'].fillna(value = tempDF["X7"].mean(), inplace = True)
tempDF['X8'].fillna(value = tempDF["X8"].mean(), inplace = True)
tempDF['X9'].fillna(value = tempDF["X9"].mean(), inplace = True)
tempDF['X10'].fillna(value = tempDF["X10"].mean(), inplace = True)
tempDF['X11'].fillna(value = tempDF["X11"].mean(), inplace = True)
tempDF['X12'].fillna(value = tempDF["X12"].mean(), inplace = True)
tempDF['X13'].fillna(value = tempDF["X13"].mean(), inplace = True)
tempDF['X14'].fillna(value = tempDF["X14"].mean(), inplace = True)
tempDF['X15'].fillna(value = tempDF["X15"].mean(), inplace = True)
tempDF['X16'].fillna(value = tempDF["X16"].mean(), inplace = True)
tempDF['X17'].fillna(value = tempDF["X17"].mean(), inplace = True)
tempDF['X18'].fillna(value = tempDF["X18"].mean(), inplace = True)
tempDF['X19'].fillna(value = tempDF["X19"].mean(), inplace = True)
tempDF['X20'].fillna(value = tempDF["X20"].mean(), inplace = True)
tempDF['X21'].fillna(value = tempDF["X21"].mean(), inplace = True)
tempDF['X22'].fillna(value = tempDF["X22"].mean(), inplace = True)
tempDF['X23'].fillna(value = tempDF["X23"].mean(), inplace = True)


tempDF['X2'].fillna(value = tempDF["X2"].median(), inplace = True)
tempDF['X3'].fillna(value = tempDF["X3"].median(), inplace = True)
tempDF['X4'].fillna(value = tempDF["X4"].median(), inplace = True)

print("Total de Registros después de eliminar los nulos: "+ str(len(tempDF)))
tempDF.isnull().any() #Se checa si el dataframe aún tiene registros nulos
```

Con este método conservamos todos los registros llenando el espacio de los valores nulos

```python
tempDF.rename( columns = { "X1": "Credit_Limit",
                          "X2": "Gender",
                          "X3": "Education",
                          "X4": "Marital_Status",
                          "X5": "Age"
                          }, inplace = True)
tempDF.head()
```

Se despliega la matriz de correlación para saber cuálaes datos tienen mejor acoplamiento.

```python
#Se importan parámetros y se asigna un tamaño de gráfica
from matplotlib import rc_params
sns.set(rc = {"figure.figsize":(20,14)})

#se esxtraen los valores de correlación del dataframe
matriz = tempDF.corr()
#Se confifgura la gráfica con los datos, el tipo de color y los valores minimos y maximos
plt.matshow(matriz, cmap = "BuGn", vmin = -1, vmax = 1)
plt.xticks(range(25), tempDF.columns, rotation = 90)
plt.yticks(range(25), tempDF.columns)

#Se hace dos ciclos anidados para desplegar los valores de acomplamiento.
for r in range(len(tempDF.columns)):
  for c in range(len(tempDF.columns)):
    plt.text(r,c,round(matriz.iloc[r,c],1), ha = "center", va = "center")

#Se despliega la barra de colores
plt.colorbar()
#Se despliega la gráfica
plt.show()
```

Resulta que los datos con mejor acoplamiento son los de "Edad" con "Estado Civil"
